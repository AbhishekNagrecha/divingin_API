```python
import requests
```

```python
response = requests.get("https://api.covid19api.com/summary").text
```

```python
response
```

```python
# Creating a python dictionary from json object

import json
response_information = json.loads(response)
response_information
```

```python
type(response_information)
```

```python
#  looping through the values of the ‘Countries’

response_information['Countries']
```

```python
country_list = []
for country_information in response_information["Countries"]:
     country_list.append([country_information["Country"], country_information["TotalRecovered"]])
```

```python
country_list
```

```python
country_list2 = []
for country_information in response_information["Countries"]:
     country_list2.append([country_information["Country"], country_information["TotalConfirmed"]])
```

```python
country_list2
```

```python
# Now creating a pandas DataFrame

import pandas as pd
country_df = pd.DataFrame(data=country_list, columns=["Country", "Total_Recovered"])
country_df.head(8)
```

```python
# Now creating a pandas DataFrame

import pandas as pd
country_df2 = pd.DataFrame(data=country_list2, columns=["Country", "Total_Confirmed"])
country_df2.head(8)
```
